```python
import numpy as np
import csv
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
with open('wengrae_rewards_firstrun.csv', 'r') as f:
    reader = csv.reader(f)
    raw_coins = list(reader)
coins = np.array([item for sublist in raw_coins for item in sublist]).astype(np.float)
print(np.shape(coins))
```

```python
rounds = np.arange(len(coins))
```

```python
plt.plot(rounds,coins,linestyle='None',marker='x')
```

```python
plt.plot(rounds[:1000],coins[:1000],linestyle='None',marker='x')
```

```python
plt.plot(rounds[8000:9000],coins[8000:9000],linestyle='None',marker='x')
```

```python
N = 5
y, z = np.zeros(N), np.zeros(N)
x = np.arange(N)*100
for i in range(N):
    start = i*100
    stop = (i+1)*100
    sample = coins[start:stop]
    y[i] = np.mean(sample)
    z[i] = np.sum(sample[sample==9])
plt.plot(x,y)
#plt.ylim(0,8)
plt.title('Agent performance in final training')
plt.xlabel('Number of trained games')
plt.ylabel('Average number of coins collected per game')
plt.savefig('final_training_average.pdf')
```

```python
with open('wengrae_rewards.csv', 'r') as f:
    reader = csv.reader(f)
    raw_coins = list(reader)
coins = np.array([item for sublist in raw_coins for item in sublist]).astype(np.float)
print(np.shape(coins))
```

```python
N = 5
y, z = np.zeros(N), np.zeros(N)
x = np.arange(N)*100
for i in range(N):
    start = i*100
    stop = (i+1)*100
    sample = coins[start:stop]
    y[i] = np.mean(sample)
    z[i] = np.sum(sample[sample==9])
plt.plot(x,y)
plt.ylim(0,8)
plt.title('Agent performance on first task (2 layer network)')
plt.xlabel('Number of trained games')
plt.ylabel('Average number of coins collected per game')
plt.savefig('final_training_average_2.pdf')
```
