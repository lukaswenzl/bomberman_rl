```python
v2 = np.loadtxt("agent_step1_v2_rewards.csv")
v3 = np.loadtxt("agent_step1_v3_rewards.csv")
v4 = np.loadtxt("agent_step1_v4_rewards.csv")
```

```python
#plt.plot(range(len(v2)), v2, ".")
#plt.plot(range(len(v3)), v3, ".")
#plt.plot(range(len(v4)), v4, ".")

plt.figure(figsize=(5,3))

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

plt.plot(range(len(v2)), smooth(v2,20), label="step1_v2")
plt.plot(range(len(v3)), smooth(v3,20), label="step1_v3")
plt.plot(range(len(v4)), smooth(v4,20), label="step1_v4")
plt.legend(loc="upper left")
plt.xscale("log")
plt.xlabel("number of games")
plt.ylabel("reward averaged over 20 games")
plt.savefig("step1_v2_to_v4_convergence.pdf")
```
